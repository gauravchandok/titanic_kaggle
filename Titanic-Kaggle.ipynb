```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file1= 'train.csv'
data_train= pd.read_csv(file1)
```

```python
data_train.head()
```

```python
data_train
```

```python
data_train.isnull().sum()
```

```python
data_train= data_train.drop(['Age','Cabin'], axis= 1)
data_train= data_train.dropna()
```

```python
data_survived= data_train.Survived.value_counts()
data_survived
```

```python
data_sex= data_train.Sex.value_counts()
data_sex
```

```python
data_male= data_train.Survived[data_train.Sex== 'male'].value_counts().sort_index()
data_male
```

```python
data_female= data_train.Survived[data_train.Sex== 'female'].value_counts().sort_index()
data_female
```

```python
data_class_survived= data_train.Pclass[data_train.Survived== 1].value_counts()
data_class_died= data_train.Pclass[data_train.Survived== 0].value_counts()
print (data_class_survived)
print (data_class_died)
```

```python
graph= plt.figure(figsize=(4,5))

data_male.plot(kind='bar', color= "blue", label= 'Male', alpha=0.55)
data_female.plot(kind='bar', color= "pink", label= 'Female', alpha=0.55)
plt.title("Survived vs Died (1= Survived 0= Died)"); plt.legend(loc='best')
```

```python
graph2= plt.figure(figsize=(4,5))

data_male_class3= data_train.Survived[data_train.Sex== 'male'][data_train.Pclass== 3].value_counts().sort_index()
data_male_class3.plot(kind='bar', color="brown", label= "Class 3" , alpha= 0.65)
plt.title("Male Survived vs Died (1= Survived 0= Died)");plt.legend (loc='best')
```

```python
graph3= plt.figure(figsize= (4,5))

data_male_class2= data_train.Survived[data_train.Sex== "male"][data_train.Pclass== 2].value_counts().sort_index()
data_male_class2.plot(kind='bar', color= "silver", label= "Class 2", alpha= 0.65)
plt.title("Male Survived vs Died (1= Survived 0= Died)"); plt.legend (loc='best')
```

```python
graph4= plt.figure(figsize= (4,5))

data_male_class1= data_train.Survived[data_train.Sex== "male"][data_train.Pclass== 1].value_counts().sort_index()
data_male_class1.plot(kind='bar', color= "gold", label= "Class 1", alpha= 0.65)
plt.title("Male Survived vs Died (1= Survived 0= Died)"); plt.legend (loc='best')
```
